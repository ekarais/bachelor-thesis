```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
root_dir = "/home/ege/Documents/bthesis/data/"
data_dir = "orig_data/neutral/"
filename = "PGM_neutral_train_145921.npz"
```

```python
pgm = np.load(root_dir+data_dir+filename)
pgm.shape
```

```python
pgm = pgm["image"].reshape(16, 160, 160)
```

```python
c = np.ravel_multi_index(([1],[1]),(3,3))
c[0]
```

```python
fig, ax = plt.subplots(3, 3, sharex='col', sharey='row')
fig.set_size_inches(10.5, 10.5)

for i in range(3):
    for j in range(3):
        c = np.ravel_multi_index(([i],[j]),(3,3))[0]
        if not (i==2 and j==2):
            ax[i, j].imshow(pgm[c], cmap='binary')
            ax[i, j].title.set_text("Panel {}".format(c))
```
